```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from Environment_1D import Environment
size = 100 # size of the map in meters
goal_range = [40,45]
goal_reward = 100. # reward for reaching goal
punishment = -100. # punishment for going out of bounds
time_penalty = -1.
timestep = .1 # seconds
env = Environment(size,goal_range,goal_reward,punishment,time_penalty,timestep)
```

```python
from Agents import DDPGAgent
# The State give the Cartesian coordinates for both Position and Velocity
# State: [x,v]
state_dim = 2
# The action is the proposed acceleration
# Action: a
action_dim = 1
max_action = 1
min_action = -1
agent = DDPGAgent(state_dim, action_dim, min_action, max_action)

from Engine import Engine
engine = Engine(env,agent)
```

```python
avg_accel = []
for i in range(100):
    states,actions,next_states,rewards = engine.run(init_state = [i,0],gaussian_noise = .0)
    avg_accel.append(actions.mean() if len(actions) else 0.)
plt.plot(avg_accel)
plt.gca().axvspan(40,45,color='gold',alpha=.4)
plt.ylim(-1.1,1.1)
```

```python
metrics = engine.train(1000, init_state = [35,0], gaussian_noise = .1, limit = 1000)
df = pd.DataFrame(metrics)
fix,ax = plt.subplots(3,1,figsize=(10,15))
sns.lineplot(df.reward,ax=ax[0])
ax[0].set_title('Rewards')
sns.lineplot(df.actor_loss,ax=ax[1])
ax[1].set_title('Actor Loss')
sns.lineplot(df.critic_loss,ax=ax[2])
ax[2].set_title('Critic Loss');
```
